### RiseSet Calendar

#### A unique calendar writen in Python with solarsystem package which provides:
* Time of sun rise and set within each day
* Time of moon rise and set within each day
* Moon phase - percent of illumination

```python
import solarsystem
import datetime
```

```python
print(solarsystem.__version__)
```

```python
def datetime2timestring(dt):
    '''
    Helper Function to read a datetime and retrun the time as a string of form hour:minute
    '''
    h=str(dt.hour)
    if len(h)==1: h = '0' + h
    m=str(dt.minute)
    if len(m)==1: m = '0' + m
    return h+':'+m
```

```python
year=2026

# Athens Greece
longtitude = 23.72
latitude   = 37.98
# Athens is in GMT +2 time zone (setting UT = 2)
UT = 2

#########################
#Don' t edit time below
hour=12
minute=0
```

```python
week = ['Mo', 'Tu', 'We', 'Th', 'Fr', 'Sa', 'Su']
```

```python
# For Europe DST starts on last Sunday of March and Ends on last Sunday of October
for i in range(31,31-7,-1):
    daydate = datetime.datetime(2020, 3, i)
    strday=week[daydate.weekday()]
    if strday == 'Su':
        dst_starts = i
        break
for i in range(31,31-7,-1):
    daydate = datetime.datetime(2020, 10, i)
    strtends=week[daydate.weekday()]
    if strtends == 'Su':
        dst_ends = i
        break       
print('DST starts:', dst_starts, 'March  and DST ends: ', dst_ends, 'of October' )
```

```python
# Helping dictionaries and lists
calender = [('January', range(1, 31 + 1)),
            ('Feburary', range(1, 28 + 1)),
            ('March', range(1, 31 + 1)),
            ('April', range(1, 30 + 1)),
            ('May', range(1, 31 + 1)),
            ('June', range(1, 30 + 1)),
            ('July', range(1, 31 + 1)),
            ('August', range(1, 31 + 1)),
            ('September', range(1, 30 + 1)),
            ('October', range(1, 31 + 1)),
            ('November', range(1, 30 + 1)),
            ('December', range(1, 31 + 1))]

month2num = {'January': 1,
            'Feburary': 2,
            'March': 3,
            'April': 4,
            'May': 5,
            'June': 6,
            'July': 7,
            'August': 8,
            'September': 9,
            'October': 10,
            'November': 11,
            'December':12,
           }
daysinmonth = {'January': 31,
            'Feburary': 28,
            'March': 31,
            'April': 30,
            'May': 31,
            'June': 30,
            'July': 31,
            'August': 31,
            'September': 30,
            'October': 31,
            'November': 30,
            'December':31,
           }
```

### Moon rise may be after a moon set within a day. In cases where in a calendar day no moon rise/set takes place no time is written. In these cases you can search for the moorise in the previous day and for the moon set in the following day.

```python
def is_leap(year):
    """Checks if year is a leap year"""
    if year % 4 == 0:
        if year % 100 == 0:
            if year % 400 == 0:
                return True
            else:
                return False
        else:
            return True
    else:
        return False

def make_calendar(year, start_day):
    """
    make_calendar(int, str) --> None
    """
    # Determine current starting position on calendar
    start_pos = week.index(start_day)

    # if True, adjust Feburary date range for leap year | 29 days
    if is_leap(year):
        calender[1] = ('Feburary', range(1, 29 + 1))
        daysinmonth['Feburary'] = 29
        
    for month, days in calender:

            
        # Print month title
        print('{0} {1}'.format(month, year).center(20, ' '))
        # Print Day headings
        print(''.join(['{0:<13}'.format(w) for w in week]))
        # Add spacing for non-zero starting position
        print('{0:<13}'.format('')*start_pos, end='')
        start_pos2=start_pos
        rss=''
        rss2=''
        rss2phase=''
        for day in days:
            is_dst = 0
        
            if (month2num[month]>3)&(month2num[month]<10):
                is_dst = 1
            if ((month2num[month]==3)&(day>dst_starts))|((month2num[month]==10)&(day<dst_ends)):
                is_dst = 1
            
            # Print day
            print('{0:<13}'.format(day), end='')
            sun = solarsystem.Sunriseset(year=year, month=month2num[month], day=day, 
                         UT=UT, dst=is_dst, longtitude=longtitude, latitude=latitude)

#             rss = (rss + solarsystem.demical2clock(sun.riseset()[0])[:-3]+'-'+
#                    solarsystem.demical2clock(sun.riseset()[1])[:-3] + '  ')
            rss = (rss +datetime2timestring(sun.riseset()[0])+'|'+datetime2timestring(sun.riseset()[1])+ '  ')
            
            moon = solarsystem.Moon(year=year, month=month2num[month], day=day, hour=hour, minute=minute, 
                         UT=UT, dst=is_dst, longtitude=longtitude, latitude=latitude, topographic=True)
#             rss2 = (rss2 + solarsystem.demical2clock(moon.moonriseset()[0])[:-3]+'-'+
#                    solarsystem.demical2clock(moon.moonriseset()[1])[:-3] + '  ')
            if moon.moonriseset()[0].day!=day:
                rss2 = (rss2 + '     |'+datetime2timestring(moon.moonriseset()[1]) + '  ')
            elif  moon.moonriseset()[1].day!=day:
                rss2 = (rss2 + datetime2timestring(moon.moonriseset()[0])+'|     ' + '  ')
            else:
                rss2 = (rss2 + datetime2timestring(moon.moonriseset()[0])+'|'+
                               datetime2timestring(moon.moonriseset()[1]) + '  ')
            rss2phase = rss2phase + str(round(moon.phase()*100))  + '%          '
            if len(str(round(moon.phase()*100)))==1:
                rss2phase=rss2phase+' '
            if len(str(round(moon.phase()*100)))==3:
                rss2phase=rss2phase[:-1]               
            
            start_pos += 1
            if (start_pos == 7)|(day==daysinmonth[month]):
                # If start_pos == 7 (Sunday) start new line
                print()
                if (start_pos == 7):
                    start_pos = 0 # Reset counter
                if (start_pos == 0)&(day<=7):
                    rss = '{0:<13}'.format('')*start_pos2+rss
                    rss2 = '{0:<13}'.format('')*start_pos2+rss2
                    rss2phase = '{0:<13}'.format('')*start_pos2+rss2phase
                print(rss)
                print(rss2)
                print(rss2phase)
                rss=''
                rss2=''
                rss2phase=''
                print()
                

                
        print('\n')


    
daydate11 = datetime.datetime(year, 1, 1)
strtday=week[daydate11.weekday()]
make_calendar(year,strtday)
```
